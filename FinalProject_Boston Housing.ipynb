## Task 3: Apply Statistical Tests

1. Use Levene’s test to check if the variances of MEDV for homes bounded by the Charles River (CHAS = 1) and those not bounded (CHAS = 0) are equal.

2. Perform a T-test to determine if there is a significant difference in MEDV between homes bounded and not bounded by the Charles River.

```python
#1. Lets create a bar plot to see how MEDV is affected by bounding of the Charles River. 

sns.histplot(data = df, x = "MEDV", hue = "CHAS")
```

In this histogram we can see that the MEDV for housing not bounded by the Charles river seem to have a wider spread (a higher variance).

```python
# 1. Use Levene’s test to check if the variances of MEDV for homes bounded by the Charles River (CHAS = 1) and those not bounded (CHAS = 0) are equal.

scipy.stats.levene(df[df["CHAS"] == 0]["MEDV"], df[df["CHAS"] == 1]["MEDV"], center = "mean")
```

We state the hypothesis for this Levene's test: 

1. $H_0$ : There is homocedasticity, $\sigma_1 = \sigma_2$.
2. $H_a$ : There is heterocedasticity, $\sigma_1 \neq \sigma_2$.

For $\alpha = 0.05$ we have $ P < \alpha $ therefore we reject the null hypothesis, as there is statistically significant evidence that there is heterocedasticity.

```python
#2. Let's create a boxplot to see how MEDVs compare for houses bounded and not bounded by the Charles River.

ax = sns.boxplot(x = "CHAS", y = "MEDV", data = df)
ax.set(xlabel = "Bounded by the Charles River, 0 = no, 1 = yes", ylabel = "Median Value", title = "Value of houses per bounding of the Charles River")
```

In the previous boxplot we can get the intuition that housing bounded by the Charles River have higher median values, however, houses not bounded by the river have more outliers with high values.

```python
# 2. Perform a T-test to determine if there is a significant difference in MEDV between homes bounded and not bounded by the Charles River.

scipy.stats.ttest_ind(df[df["CHAS"] == 0]["MEDV"], df[df["CHAS"] == 1]["MEDV"], equal_var = True)
```

We state the hypothesis for the t-test:

1. $H_0$: Poblational means are equal, $\mu_1 = \mu_2$.
2. $H_a$: Poblational means differ, $\mu_1 \neq \mu_2$.

For $\alpha = 0.05$ we have $ P << \alpha $ therefore we reject the null hypothesis, as there is statistically significant evidence that the poblational means are different.

```python
x_hat_0 = df[df["CHAS"] == 0]["MEDV"].mean()
x_hat_1 = df[df["CHAS"] == 1]["MEDV"].mean()
print(f"The mean MEDV for houses bounded and non-bounded by the Charles River are {x_hat_1} and {x_hat_0}, respectivelly") 
```

With this info we can conclude with $95\%$ confidence that houses bounded by the Charles River have a higher MEDV.

```python
# 3. Conduct an ANOVA test to compare MEDV among the three AGE groups created earlier.

# First we split into three categories and assign it to variables

less_36 = df[df["AGE_group"] == "36 years and less"]["MEDV"]
middle = df[df["AGE_group"] == "between 36 and 68 years"]["MEDV"]
more_68 = df[df["AGE_group"] == "68 years and more"]["MEDV"]
```

```python
ax = sns.boxplot(x = "AGE_group", y = "MEDV", data = df)
ax.set(title = "Median value of houses vs age", xlabel = "Age", ylabel = "Median value")
```

In this boxplot it is visible that houses with  36 years and less usually have higher median values. It can also be noted that houses with 68 years and more tend to have lower median values. We can infer intuitively that the lower the age, the higher the MEDV.

On the spread of the values, we can see that the older houses seem to have the widest spread, while the younger houses have the narrowest spread. 

```python
# Secondly, we will perform a Levene's test to see if there is homocedasticity and we can compare population means through an ANOVA.

scipy.stats.levene(less_36, middle, more_68, center = "mean")
```

We state the hypothesis for this Levene's test: 

1. $H_0$ : There is homocedasticity, $\sigma_1 = \sigma_2 = \sigma_3$.
2. $H_a$ : There is heterocedasticity, i.e. at least one poblational variance differs.

For $\alpha = 0.05$ we have $ P > \alpha $ therefore we fail to reject the null hypothesis, as there is statistically significant evidence that there is homocedasticity, which allows us to test for poblational means differences through an ANOVA test.

We state the hypothesis for this one way ANOVA test: 

1. $H_0$ : Poblational means are equal, $\mu_1 = \mu_2 = \mu_3$.
2. $H_a$ : At least one poblational mean differs.

```python
f_statistic, p_value = scipy.stats.f_oneway(less_36, middle, more_68)
print(f"F_statistic {f_statistic}, P-value: {p_value}")
```

For $\alpha = 0.05$ we have $ P << \alpha $ therefore we reject the null hypothesis, as there is statistically significant evidence that the poblational means are different.

```python
m1 = less_36.mean()
m2 = middle.mean()
m3 = more_68.mean()

print(f"The means MEDV for houses for younger, middle and older houses are {m1}, {m2} and {m3}, respectivelly.")
print("With this information and the results of the one-way ANOVE we conclude that younger houses have higher MEDVs.")
```

```python
# 4. Perform a Pearson correlation test to assess the relationship between NOX and INDUS.

scipy.stats.pearsonr(df["INDUS"], df["NOX"])
```

We state the hypothesis for the Pearson correlation test:

*   $H_0:$ Nitric oxide concentration is not correlated with non-retail business acres.
*   $H_1:$ Nitric oxide concentration is correlated with non-retail business acres.

For $\alpha = 0.05$ we have $ P << \alpha $ therefore we reject the null hypothesis, as we are 95% confident that INDUS and NOX are correlated. As our $ R = 0.76 $ we have a strong positive relationship, which is to say that as INDUS increases, NOX increases.

```python
# 5. Fit a simple linear regression model to evaluate the impact of DIS (distance to employment centres) on MEDV. 
# Report and interpret the regression results.

X = df["DIS"]
y = df["MEDV"]
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
y_hat = model.predict(X)
model.summary()
```

## Task 3: Apply Statistical Tests

1. Use Levene’s test to check if the variances of MEDV for homes bounded by the Charles River (CHAS = 1) and those not bounded (CHAS = 0) are equal.

2. Perform a T-test to determine if there is a significant difference in MEDV between homes bounded and not bounded by the Charles River.

3. Conduct an ANOVA test to compare MEDV among the three AGE groups created earlier.

4. Perform a Pearson correlation test to assess the relationship between NOX and INDUS.

5. Fit a simple linear regression model to evaluate the impact of DIS (distance to employment centres) on MEDV. Report and interpret the regression results.

```python
# 1. Use Levene’s test to check if the variances of MEDV for homes bounded by the Charles River (CHAS = 1) and those not bounded (CHAS = 0) are equal.


scipy.stats.levene(df[df["CHAS"] == 0]["MEDV"], df[df["CHAS"] == 1]["MEDV"], center = "mean")
```

```python
# 2. Perform a T-test to determine if there is a significant difference in MEDV between homes bounded and not bounded by the Charles River.


scipy.stats.ttest_ind(df[df["CHAS"] == 0]["MEDV"], df[df["CHAS"] == 1]["MEDV"], equal_var = True)
```

```python
sns.boxplot(x = "CHAS", y = "MEDV", data = df)
```

```python
# 3. Conduct an ANOVA test to compare MEDV among the three AGE groups created earlier.

less_36 = df[df["AGE_group"] == "36 years and less"]["MEDV"]
middle = df[df["AGE_group"] == "between 36 and 68 years"]["MEDV"]
more_68 = df[df["AGE_group"] == "68 years and more"]["MEDV"]

scipy.stats.levene(less_36, middle, more_68, center = "mean")
```

```python
sns.boxplot(x = "AGE_group", y = "MEDV", data = df)
```

```python
f_statistic, p_value = scipy.stats.f_oneway(less_36, middle, more_68)
print(f"F_statistic {f_statistic}, P-value: {p_value}")
```

```python
# 4. Perform a Pearson correlation test to assess the relationship between NOX and INDUS.

scipy.stats.pearsonr(df["INDUS"], df["NOX"])
```

```python
# 5. Fit a simple linear regression model to evaluate the impact of DIS (distance to employment centres) on MEDV. 
# Report and interpret the regression results.

X = df["DIS"]
y = df["MEDV"]
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
y_hat = model.predict(X)
model.summary()
```

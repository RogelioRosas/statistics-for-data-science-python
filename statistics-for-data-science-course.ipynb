# Notebook 5: Hypothesis testing

The goal of hypothesis testing is to answer the question, “Given a sample and an apparent effect, what is the probability of seeing such an effect by chance?” The first step is to quantify the size of the apparent effect by choosing a test statistic (t-test, ANOVA, etc). The next step is to define a null hypothesis, which is a model of the system based on the assumption that the apparent effect is not real. Then compute the p-value, which is the probability of the null hypothesis being true, and finally interpret the result of the p-value, if the value is low, the effect is said to be statistically significant, which means that the null hypothesis may not be accurate.

### T-Test: Using the teachers' rating data set, does gender affect teaching evaluation rates?

We will be using the t-test for independent samples. For the independent t-test, the following assumptions must be met.

*   One independent, categorical variable with two levels or group
*   One dependent continuous variable
*   Independence of the observations. Each subject should belong to only one group. There is no relationship between the observations in each group.
*   The dependent variable must follow a normal distribution
*   Assumption of homogeneity of variance

State the hypothesis

*   $H_0: µ_1 = µ_2$ ("there is no difference in evaluation scores between male and females")
*   $H_1: µ_1 ≠ µ_2$ ("there is a difference in evaluation scores between male and females")


We can plot the dependent variable with a histogram to know if the scores follow a normal distribution.


```python
sns.histplot(df["eval"], bins = 20, kde = True, color = "green")
ax.set(xlabel = "Normal Distribution", ylabel = "Frequency")
plt.show()

# it seems to follow a normal distribution
```

```python
# Let's check the homogeneity of variance through a Levene's test
```

```python
scipy.stats.levene(df[df["gender"] == "female"]["eval"],
                   df[df["gender"] == "male"]["eval"], center = "mean")
```

$ p > 0.05 $ so we can asume $s_1 = s_2$

```python
# Let's test if the mean of both populations is equal

scipy.stats.ttest_ind(df[df["gender"] == "female"]["eval"],
                      df[df["gender"] == "male"]["eval"], equal_var = True)
```

Since $p < 0.05$ we reject the null hypothesis, so there is a difference in teaching evaluatios due to gender. 

```python
sns.boxplot(x = "gender", y = "eval", data = df)
plt.show()
```

### ANOVA: Using the teachers' rating data set, does beauty  score for instructors  differ by age?

First, we group the data into cateries as the one-way ANOVA can't work with continuous variable - using the example from the video, we will create a new column for this newly assigned group our categories will be teachers that are:

*   40 years and younger
*   between 40 and 57 years
*   57 years and older



```python
# Creating bins for the three different age groups

df.loc[(df["age"] <= 40), "age_group"] = "40 years and younger"
df.loc[(df["age"] > 40)&(df["age"] < 57), "age_group"] = "between 40 and 57 years"
df.loc[(df["age"] >= 57), "age_group"] = "57 years and older"
```

```python
df.head()
```

State the hypothesis


*   $H_0: µ_1 = µ_2 = µ_3$ (the three population means are equal)
*   $H_1:$ At least one of the means differ

Test for equality of variance


```python
scipy.stats.levene(df[df["age_group"] == "40 years and younger"]["beauty"],
                   df[df["age_group"] == "between 40 and 57 years"]["beauty"], 
                   df[df["age_group"] == "57 years and older"]["beauty"], center = "mean")
```

$ p < 0.05 $ so variances are not equal, however, for the purposes of the exersice, we move forward without further consideration.

First, separate the three samples (one for each job category) into a variable each.


```python
younger_40 = df[df["age_group"] == "40 years and younger"]["beauty"]
middle_40_57 = df[df["age_group"] == "between 40 and 57 years"]["beauty"]
older_57 = df[df["age_group"] == "57 years and older"]["beauty"]
```

Now, run a one-way ANOVA.


```python
f_statistic, p_value = scipy.stats.f_oneway(younger_40, middle_40_57, older_57)
print(f"F_statistic {f_statistic}, P-value: {p_value}")
```

As $ p < 0.05 $ we reject the null hypothesis, so at least one of the means differ. 

### ANOVA: Using the teachers' rating data set, does teaching  evaluation  score for instructors  differ  by age?


```python
# testing for equality of variances

scipy.stats.levene(df[df["age_group"] == "40 years and younger"]["eval"],
                   df[df["age_group"] == "between 40 and 57 years"]["eval"], 
                   df[df["age_group"] == "57 years and older"]["eval"], center = "mean")
```

$ p < 0.05 $ so variances are not equal, however, for the purposes of the exersice, we move forward without further consideration.

```python
eval_younger_40 = df[df["age_group"] == "40 years and younger"]["eval"]
eval_middle_40_57 = df[df["age_group"] == "between 40 and 57 years"]["eval"]
eval_older_57 = df[df["age_group"] == "57 years and older"]["eval"]
```

```python
f_statistic, p_value = scipy.stats.f_oneway(eval_younger_40, eval_middle_40_57, eval_older_57)
print(f"F_statistic {f_statistic}, P-value: {p_value}")
```

$ p > 0.05 $ so we fail to reject the null hypothesis, that means there is no statistically significant evidence that means differ across the three groups, i.e. teaching evaluation is not affected by the age.

### Chi-square contingency test: Using the teachers' rating data set, is there an association between tenure and gender?


State the hypothesis:

*   $H_0:$ The proportion of teachers who are tenured is independent of gender
*   $H_1:$ The proportion of teachers who are tenured is associated with gender

I expect that there is an association between gender and tenure, as our graphs of the data show that there are considerably more tenured male teachers.

```python
# To proof for association between labeled data in two labels (gender, tenure) we use the chi-square test, and it needs a cross-tab table.
```

```python
cross_t = pd.crosstab(df["tenure"], df["gender"])
cross_t
```

```python
# We apply the chi2 contingency test, without the Yates correction because our sample are sufficiently big (frequencies bigger than 10)

chi2, pvalue, dof, exp_tab = scipy.stats.chi2_contingency(cross_t, correction = False)
print(f"The chi2 statistic is: {chi2}, P-value: {pvalue}")
```

$ p > 0.05 $ so we fail to reject the null hypothesis, therefore there is not enough evidence to conclude that teacher tenure is associated with gender.

```python
# Question 1: Since we want to compare a continuos dependet variable (evaluation) with a categorical independent variable we (tenure) use a T-test.

# We need to check that the continuos variable is normally distributed. 
# We need to check that there is homogeneity of variance in the dependent variable categories.

# For the normal distribution, let's use normality tests for non-standard normal distributions

from scipy.stats import shapiro, normaltest

# Shapiro-Wilk test
stat, p_value_shapiro = shapiro(df["eval"])
print(f"Shapiro-Wilk Statistic: {stat}, p-value: {p_value_shapiro}")

# D'Agostino-Pearson K^2 test
stat, p_value_normaltest = normaltest(df["eval"])
print(f"Normaltest Statistic: {stat}, p-value: {p_value_normaltest}")
```

### Question 2: Using the teachers rating data set, is there an association between age and tenure?

*   Discretize the age into three groups 40 years and youngers, between 40 and 57 years, 57 years and older (This has already been done for you above.)
*   What is your conclusion at α = 0.01 and α = 0.05?

In this problem we have "tenure", a categorical variable and "age" a continuos variable. Here "tenure" is de dependent variable and age the indepentent variable. Because of this, we can not use a t-test, so if we change the problem to one between two categorical variables, we can use a Chi-square test.

* First we will split the age variable into three labels, 40 years and younger, between 40 and 57 yo and 57 years and older.
* Then we will perform our Chi-square test with $\alpha = 0.01$ and $\alpha = 0.05$.

We state the hypothesis: 

*   $H_0:$ The proportion of teachers who are tenured is independent of age.
*   $H_1:$ The proportion of teachers who are tenured is associated with age.


```python
df.head()
```

```python
# We split the data into the three different age bins.

df.loc[(df["age"] <= 40), "age_group"] = "40 years and younger"
df.loc[(df["age"] > 40)&(df["age"] < 57), "age_group"] = "between 40 and 57 years"
df.loc[(df["age"] >= 57), "age_group"] = "57 years and older"
```

```python
# Now we assign a variable to each group

ten_younger_40 = df[df["age_group"] == "40 years and younger"]["tenure"]
ten_middle_40_57 = df[df["age_group"] == "between 40 and 57 years"]["tenure"]
ten_older_57 = df[df["age_group"] == "57 years and older"]["tenure"]
```

```python
sns.boxplot(x = "age", y = "tenure", data = df)
plt.show()
```

```python
sns.catplot(x = "age_group", hue = "tenure", kind = "count", data = df)
plt.show()
```

```python
# We get the crosstab for the chi-square test 

crosst2 = pd.crosstab(df["age_group"], df["tenure"])
crosst2
```

```python
# We apply the chi2 contingency test, without the Yates correction because our sample are sufficiently big (frequencies bigger than 10)

chi2, pvalue, dof, exp_tab = scipy.stats.chi2_contingency(crosst2, correction = False)
print(f"The chi2 statistic is: {chi2}, P-value: {pvalue}")
```

If $\alpha = 0.05 \rightarrow p < \alpha$, so we reject the null hypothesis and then there is an association between age and tenure, however
if $\alpha = 0.01 \rightarrow p > \alpha$, so we fail to reject the null hypothesis and then tenure is independent of age. 

This means that we can be $95 \%$ sure that there is an association between age and tenure, but we can not be $99\%$ sure that this association exists.

# **Regression Analysis**


The goal of regression analysis is to describe the relationship between one set of variables called the dependent variables, and another set of variables, called independent or explanatory variables. When there is only one explanatory variable, it is called simple regression.

```python
import pandas as pd
import numpy as np
import scipy.stats
import statsmodels.api as sm
```

```python
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/teachingratings.csv"
df = pd.read_csv(URL)
```

## Lab Exercises

In this section, you will learn how to run regression analysis in place of the t-test, ANOVA, and correlation.

### Regression with T-test: Using the teachers rating data set, does gender affect teaching evaluation rates?

Initially, we had used the t-test to test if there was a statistical difference in evaluations for males and females, we are now going to use regression. We will state the null hypothesis:
* $H_0: β1$ = 0 (Gender has no effect on teaching evaluation scores)
* $H_1: β1$ is not equal to 0 (Gender has an effect on teaching evaluation scores)
  
We will use the female variable. female = 1 and male = 0



```python
df.head()
```

```python
## X is the input variables (or independent variables)
X = df['female']
## y is the target/dependent variable
y = df['eval']
## add an intercept (beta_0) to our model
X = sm.add_constant(X) 

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

# Print out the statistics
model.summary()
```

**Conclusion:** Like the t-test, the p-value is less than the alpha (α) level = 0.05, so we reject the null hypothesis as there is evidence that there is a difference in mean evaluation scores based on gender. The coefficient -0.1680 means that females get 0.168 scores less than men.


### Regression with ANOVA: Using the teachers' rating data set, does beauty  score for instructors  differ by age?


State the Hypothesis:
* $H_0: µ_1 = µ_2 = µ_3$ (the three population means are equal)
* $H_1:$ At least one of the means differ

```python
df.loc[(df['age'] <= 40), 'age_group'] = '40 years and younger'
df.loc[(df['age'] > 40)&(df['age'] < 57), 'age_group'] = 'between 40 and 57 years'
df.loc[(df['age'] >= 57), 'age_group'] = '57 years and older'
```

```python
#Use OLS function from the statsmodel library

from statsmodels.formula.api import ols
lm = ols('beauty ~ age_group', data = df).fit()
table= sm.stats.anova_lm(lm)
print(table)
```

**Conclusion:** We can also see the same values for ANOVA like before and we will reject the null hypothesis since the p-value is less than 0.05 there is significant evidence that at least one of the means differ.

### Regression with ANOVA option 2
Create dummy variables - A dummy variable is a numeric variable that represents categorical data, such as gender, race, etc. Dummy variables are dichotomous, i.e they can take on only two quantitative values.


```python
X = pd.get_dummies(df['age_group']).astype(float) # we cast the boolean dummy values to float values so the regression model works
X.info()
```

```python
y = df['beauty']
## add an intercept (beta_0) to our model
X = sm.add_constant(X) 

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

# Print out the statistics
model.summary()
```

### Correlation: Using the teachers' rating dataset, Is teaching evaluation score correlated with beauty score?

```python
X = df["beauty"]
y = df["eval"]
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

model.summary()
```

**Conclusion:** p < 0.05 there is evidence of correlation between beauty and evaluation scores
